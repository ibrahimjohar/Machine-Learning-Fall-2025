```python
def f_score(X, y):
    """
    computes the f-score of each feature
    X: numpy array of shape (n_samples, n_features)
    y: numpy array of shape (n_samples)
    """
    X = np.array(X)
    y = np.array(y)
    
    pos = X[y == 1]
    neg = X[y == 0]
    
    n_pos = pos.shape[0]
    n_neg = neg.shape[0]
    
    f_scores = []
    
    for i in range(X.shape[1]):
        x_i = X[:, i]
        x_pos = pos[:, i]
        x_neg = neg[:, i]
        
        mean_total = np.mean(x_i)
        mean_pos = np.mean(x_pos)
        mean_neg = np.mean(x_neg)
        
        #numerator
        numerator = (mean_pos - mean_total)**2 + (mean_neg - mean_total)**2
        
        #denominator
        denom = (
            np.sum((x_pos - mean_pos)**2) / (n_pos - 1)
            + np.sum((x_neg - mean_neg)**2) / (n_neg - 1)
        )
    
        f_scores.append(numerator/denom)
    
    return np.array(f_scores)
```

```python
f_scores = f_score(X.values, y.values)

f_scores
```

```python
#create a sorted table of features by f-score
feature_names = X.columns

fscore_table = pd.DataFrame({
    "Feature": feature_names,
    "F-Score": f_scores
})

fscore_table_sorted = fscore_table.sort_values(by="F-Score", ascending=False).reset_index(drop=True)

fscore_table_sorted
```

```python
#generating feature subsets

#select top N features (paper uses 9)
N = 9
top_features = list(fscore_table_sorted["Feature"].head(N))

top_features
```

```python
feature_subsets = {}

for i in range(1, N+1):
    feature_subsets[f"Model_{i}"] = top_features[:i]
    
feature_subsets
```

Feature subsets help us find the smallest, most powerful combination of features.

especially useful in medical classification where:

- fewer features → simpler model

- simpler model → easier for clinicians

- less noise → better accuracy

##### implement training & evaluation for feature subsets
function that:
- Takes a feature subset

- Takes a train/test ratio

- Trains an SVM (RBF kernel - same as the paper)

- Returns accuracy

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

def evaluate_subset(features, custom_test_size):
    X_sub = df[features]
    y_sub = df["diagnosis"]
    
    #split
    X_train, X_test, y_train, y_test = train_test_split(X_sub, y_sub, test_size=custom_test_size, random_state=42, stratify=y_sub)
    
    #The scaler is fitted only on the training data 
    #and not influenced by the test data (to prevent data leakage).
    
    #scale
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    #train RBF SVM (default parameters)
    model = SVC(kernel="rbf", gamma="scale")
    model.fit(X_train_scaled, y_train)
    
    #predict and eval
    y_pred = model.predict(X_test_scaled)
    
    return accuracy_score(y_test, y_pred)
```

```python
#evaluate model_1 on 50/50 split
evaluate_subset(feature_subsets["Model_1"], custom_test_size=0.5)
```

```python
results_50 = {}

for i in range(1, 10):
    model_name = f"Model_{i}"
    acc = evaluate_subset(feature_subsets[model_name], custom_test_size=0.5)
    results_50[model_name] = acc
    
results_50
```

```python
results_70 = {}

for i in range(1, 10):
    model_name = f"Model_{i}"
    acc = evaluate_subset(feature_subsets[model_name], custom_test_size=0.3)
    results_70[model_name] = acc

results_70
```

##### **quick C tuning (for RBF kernel)**

this will test:
- low C -> soft margin
- high C -> hard margin
- find best generalisation

```python
C_values = [0.01, 0.1, 1, 10, 100]
results_C = {}

for C in C_values:
    model = SVC(kernel="rbf", C=C)
    model.fit(X_train_scaled, y_train)
    
    y_pred = model.predict(X_test_scaled)
    
    acc = accuracy_score(y_test, y_pred)
    results_C[C] = round(acc, 4)
    
results_C
```

- C = 0.01 → margin too soft → underfitting

- C = 0.1 & 1 → best performance (0.9561)

- C = 10 & 100 → slightly worse → overfitting starts

- Best C = 0.1 or 1

- (Both give identical accuracy; use C=1 for stability.)

##### **quick gamma tuning**
gamma controls how "curvy" the decision boundary is

```python
gamma_values = [0.0001, 0.001, 0.01, 0.1, 1, 10]
results_gamma = {}

for g in gamma_values:
    model = SVC(kernel="rbf", gamma=g)
    model.fit(X_train_scaled, y_train)
    
    y_pred = model.predict(X_test_scaled)
    
    acc = accuracy_score(y_test, y_pred)
    
    results_gamma[g] = round(acc, 4)
    
results_gamma
```

- Gamma too small (0.0001) → model becomes too smooth → underfits

- Gamma too large (1 or 10) → model becomes too wiggly → overfits

- Gamma = 0.1 → best performance (0.9561)

- Gamma = 0.01 → good (0.9474)

- Best Gamma = 0.1

##### **combined C + gamma tuning (small grid search)**
gives tuned model results

```python
C_values = [0.1, 1, 10]
gamma_values = [0.001, 0.01, 0.1]

results_grid = {}

for C in C_values:
    for g in gamma_values:
        model = SVC(kernel="rbf", C=C, gamma=g)
        model.fit(X_train_scaled, y_train)
        
        y_pred = model.predict(X_test_scaled)
        
        acc = accuracy_score(y_test, y_pred)
        
        results_grid[(C, g)] = acc

results_grid
```

best combination in your grid is:

- C = 10, Gamma = 0.01

with accuracy:

- 0.9649122807017544 ≈ 96.49%

```python
#compare tuned vs default
default_model = SVC(kernel="rbf")
default_model.fit(X_train_scaled, y_train)

default_default_acc = accuracy_score(y_test, default_model.predict(X_test_scaled))

default_default_acc
```
