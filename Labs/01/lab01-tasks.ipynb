#### task 03

A. Perform Graphical Visualization

B. Display a correlation graph

C. Display the distribution of Age and Frequent Cold

D. Find out outliers in Obesity feature by using Graphical Visualization

```python
#A
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(dataset['Age'], bins=20, kde=True, color="skyblue")
plt.title('distribution of age')
plt.xlabel('age')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Gender', data=dataset, palette="Set2", legend=False)
plt.title('gender distribution')
plt.xlabel('gender')
plt.ylabel('count')
plt.show()
```

```python
#B
plt.figure(figsize=(13,9))
sns.heatmap(dataset.corr(numeric_only=True), annot=False, cmap="Blues")
plt.title("correlation heatmap of dataset's features", fontsize=16)
plt.show()
```

```python
#C
plt.figure(figsize=(12,6))
sns.scatterplot(x='Age', y='Frequent Cold', data=dataset, hue='Frequent Cold', palette="Set2", alpha=0.7)

plt.title("distribution of age vs frequent cold", fontsize=14)
plt.xlabel("age")
plt.ylabel("frequent cold")
plt.yticks([0,1], ["No", "Yes"])
plt.legend(title="frequent cold")
plt.show()
```

```python
#D
plt.figure(figsize=(10,6))
sns.boxplot(x=dataset['Obesity'], color="gold")
plt.title("outlier detection in the obesity feature.", fontsize=15)
plt.xlabel("obesity")
plt.show()
```
